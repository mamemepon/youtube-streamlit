```python
from openpyxl import load_workbook, Workbook

def header_edit(ws):   # ws
    ws.cell(1, 1).value = "ご利用年月日"
    ws.cell(1, 2).value = "ご利用箇所"
    ws.cell(1, 3).value = "ご利用額"

wb1 = load_workbook("/Users/mema/Desktop/invoice.xlsx")
ws1 = wb1.active

wb2 = Workbook()
ws2 = wb2.active

header_edit(ws2)  # ws2

row_no = 2
for row in ws1.iter_rows(min_row=2):
    lst_row = [0, 0, 0]
    for cell in row:
        if cell.column == 1:  # "A"列
            lst_row[0] = cell.value
        elif cell.column == 2:  # "B"列
            lst_row[1] = cell.value
        elif cell.column == 3:  # "C"列
            lst_row[2] = cell.value
    
    for idx, val in enumerate(lst_row):
        ws2.cell(row_no, idx + 1).value = val

    row_no += 1

wb2.save("/Users/mema/Desktop/invoice1.xlsx")
```

```python
import PySimpleGUI as sg
import csv
import os
import fnmatch
from openpyxl import load_workbook
import unicodedata

def process_files():
    try:
        # CSVファイルを読み込むディレクトリとパターン
        csv_directory = './Documents/家計簿/明細書/'
        csv_pattern = 'ご利用明細_*.csv'

        # ディレクトリの内容を取得
        dir_contents = os.listdir(csv_directory)

        # Unicode正規化を行い、ファイル名を正規化
        normalized_contents = [unicodedata.normalize('NFC', f) for f in dir_contents]

        # パターンに一致するファイルを探す
        csv_files = fnmatch.filter(normalized_contents, csv_pattern)

        print("Found CSV files:", csv_files)  # デバッグ用に追加

        if not csv_files:
            raise FileNotFoundError(f"No CSV files found matching pattern: {csv_pattern}")

        csv_file_path = os.path.join(csv_directory, csv_files[0])  # 最初に見つかったファイルを使用
        data = []

        with open(csv_file_path, newline='', encoding='shift_jis') as csvfile:
            csvreader = csv.reader(csvfile)
            for row in csvreader:
                data.append(row)

        # 既存のExcelファイルを読み込む
        excel_file_path = './Documents/家計簿/明細書/inv_JRE.xlsx'
        wb = load_workbook(excel_file_path)

        # CSVファイルのファイル名をシート名にする
        sheet_name = os.path.basename(csv_file_path).replace('.csv', '')

        # 新しいシートを追加し、データを書き込む
        ws = wb.create_sheet(title=sheet_name)
        for row in data:
            ws.append(row)

        # Excelファイルを保存
        wb.save(excel_file_path)

        # 追加デバッグ: ディレクトリ内容を表示
        print("Directory contents:", normalized_contents)
        sg.popup("Success", "Data has been successfully processed and saved to the Excel file.")
    except Exception as e:
        sg.popup("Error", str(e))

# PySimpleGUIレイアウト
layout = [
    [sg.Text("CSV to Excel Processor")],
    [sg.Button("Process CSV Files")]
]

# ウィンドウの作成
window = sg.Window("CSV to Excel Processor", layout)

# イベントループ
while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED:
        break
    elif event == "Process CSV Files":
        process_files()

window.close()
```

```python
import tkinter
import csv
import os
import fnmatch
from openpyxl import load_workbook
import unicodedata

class Application(tkinter.Frame):
    def __init__(self, root=None):
        super().__init__(root, width=180, height=130, borderwidth=1, relief='groove')
        self.root = root
        self.pack()
        self.pack_propagate(0)
        self.create_widgets()

    def create_widgets(self):
        # 閉じるボタン
        quit_btn = tkinter.Button(self)
        quit_btn['text'] = '閉じる'
        quit_btn['command'] = self.root.destroy
        quit_btn.pack(side='bottom')
        
        # 実行ボタン
        exec_btn = tkinter.Button(self)
        exec_btn['text'] = '実行'
        exec_btn['command'] = self.execute_program
        exec_btn.pack(side='top')
        
    def execute_program(self):
        # CSVファイルを読み込むディレクトリとパターン
        csv_directory = './Documents/家計簿/明細書/'
        csv_pattern = 'ご利用明細_*.csv'

        # ディレクトリの内容を取得
        dir_contents = os.listdir(csv_directory)

        # Unicode正規化を行い、ファイル名を正規化
        normalized_contents = [unicodedata.normalize('NFC', f) for f in dir_contents]

        # パターンに一致するファイルを探す
        csv_files = fnmatch.filter(normalized_contents, csv_pattern)

        print("Found CSV files:", csv_files)  # デバッグ用に追加

        if not csv_files:
            raise FileNotFoundError(f"No CSV files found matching pattern: {csv_pattern}")

        csv_file_path = os.path.join(csv_directory, csv_files[0])  # 最初に見つかったファイルを使用
        data = []

        with open(csv_file_path, newline='', encoding='shift_jis') as csvfile:
            csvreader = csv.reader(csvfile)
            for row in csvreader:
                data.append(row)

        # 既存のExcelファイルを読み込む
        excel_file_path = './Documents/家計簿/明細書/inv_JRE.xlsx'
        wb = load_workbook(excel_file_path)

        # CSVファイルのファイル名をシート名にする
        sheet_name = os.path.basename(csv_file_path).replace('.csv', '')

        # 新しいシートを追加し、データを書き込む
        ws = wb.create_sheet(title=sheet_name)
        for row in data:
            ws.append(row)

        # Excelファイルを保存
        wb.save(excel_file_path)

        # 追加デバッグ: ディレクトリ内容を表示
        print("Directory contents:", normalized_contents)
        print("Excelファイルへのデータ追加が完了しました。")

root = tkinter.Tk()
root.title('inv_JREにCSV統合')
root.geometry('200x150')
app = Application(root=root)
app.mainloop()
```
